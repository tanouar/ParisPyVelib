```python
# Importation des fonctions dont l'on se servira pour toutes les figures
from bokeh.plotting import figure, output_notebook, show

# Précision de l'affichage des graphiques dans des cellules jupyter
output_notebook()
```

### II. 1 Nouveau dataframe réduit

```python
#Nouveau df avec moins de colonnes
df_lat_lon= df_hour.drop(['Unnamed: 0','Date', 'Date_instal', 'Direction', 'Y_Date_Count',
                          'M_Date_Count', 'D_Date_Count', 'Dweek_Date_Count', 'H_Date_Count',
                          'Y_Date_Instal', 'M_Date_Instal', 'D_Date_Instal', 'T°C', 'Precip_last3h',
                          'HR%', 'High_ice', 'Wind_speed_mean10mn', 'City_meteo', 'Datetime', 'nom_jour_ferie',
                          'vacances_zone_c', 'nom_vacances', 'Confinement_id'], axis =1)
#Tri par adresse
df_lat_lon = df_lat_lon.sort_values(by = ['Address'], axis=0)

df_lat_lon.head()
```

```python
df_lat_lon.nunique()
```

Pour la représentation des lieux sur une carte de Paris avec Bokeh, il faut changer les latitudes et longitudes en coordonnées mercator.
La projection de Mercator est manière de représenter les cartes géographiques en faisant une projection cylindrique.
Pour cela il est possible de créer des fonctions, adaptées de la page http://wiki.openstreetmap.org/wiki/Mercator.

```python
import math

RADIUS = 6378137.0 #rayon de la terre en mètre à l'équateur

def lat2y(a):
    return math.log(math.tan(math.pi / 4 + math.radians(a) / 2)) * RADIUS

def lon2x(a):
    return math.radians(a) * RADIUS
```

```python
df_lat_lon.info()
```

Toutes les colonnes du dataframe sont en ordre, on peut alors appliquer le changement de coordonnées.

```python
df_lat_lon['Latitude M'] = df_lat_lon['Latitude'].apply(lat2y)
df_lat_lon['Longitude M'] = df_lat_lon['Longitude'].apply(lon2x)
```

Pour une meilleure lecture avec Bokeh, il faut supprimer les apostrophes " ' ".

```python
df_lat_lon['Address'] = df_lat_lon['Address'].str.replace(r"[\"\',]", '')
```

### III.2 Représentation des données avec le module Bokeh

Tout d'abord, il faut définir les bornes de la carte pour qu'elle soit centrée autour de Paris.

```python
print('Latitude min : ', df_lat_lon['Latitude M'].min())
print('Latitude max : ', df_lat_lon['Latitude M'].max())
print('Longitude min : ', df_lat_lon['Longitude M'].min())
print('Longitude max : ', df_lat_lon['Longitude M'].max())
```

```python
#On groupe le dataFrame pour avoir le nombre de comptage total par adresse
df2 = df_lat_lon.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2 = df2.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
```

```python
from bokeh.tile_providers import get_provider
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.models.tools import HoverTool
from bokeh.models.widgets import Panel, Tabs

source = ColumnDataSource(df2)


hover = HoverTool(tooltips=[("Address", "@Address"), ("Lat", "@Latitude"), ("Lng", "@Longitude")])

p = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')


tuile = get_provider('OSM')
p.add_tile(tuile)

p.circle(x='Longitude M', y='Latitude M',source = source, fill_alpha =0.6, fill_color = 'blue',
         line_color=None, size = 15)

p.add_tools(hover)

p.title.text = "Compteurs de vélo dans Paris"
p.title.align = "center"
p.title.text_font_size = "21px"
p.title.text_color = "#333344"

show(p)
```

### III.3 Représentation des top 10

Maintenant, concentrons-nous sur les top 10 des compteurs pour les quatre années étudiées.

```python
#Top 10 des addresses avec le plus passage total
#2018
df_2018 = df_lat_lon[(df_lat_lon['Source'] == 2018)]
df2018 = df_2018.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2018 = df2018.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2018 = df2018.head(10)


#2019
df_2019 = df_lat_lon[df_lat_lon.Source == 2019]
df2019 = df_2019.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2019 = df2019.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2019 = df2019.head(10)

#2020
df_2020 = df_lat_lon[df_lat_lon.Source == 2020]
df2020 = df_2020.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2020 = df2020.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2020 = df2020.head(10)

#2021
df_2021 = df_lat_lon[df_lat_lon.Source == 2021]
df2021 = df_2021.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2021 = df2021.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2021 = df2021.head(10)

df2018['Color']= '#FECC5C'
df2019['Color']= '#FD8D3C'
df2020['Color']= '#F03B20'
df2021['Color']= '#BD0026'

#Concaténation des tops 
df_top = pd.concat([df2018, df2019, df2020, df2021])
```

```python
source = ColumnDataSource(df_top)


hover = HoverTool(tooltips=[("Address", "@Address"), ("Lat", "@Latitude"), ("Lng", "@Longitude")])

p = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')


tuile = get_provider('OSM')
p.add_tile(tuile)

p.circle(x='Longitude M', y='Latitude M',source = source, fill_alpha = 0.8, fill_color = 'Color',
         line_color=None , size = 17, legend = 'Source')

p.add_tools(hover)

p.title.text = "Top 10 des passages de vélo dans Paris selon les années"
p.title.align = "center"
p.title.text_font_size = "21px"
p.title.text_color = "#333344"

show(p)
```

En représentant les différents top selon les années, il est posssible de voir une tendance avec plus de circulation le long des bords de la Seine et le long des boulevards importants. Plus les points sont nets et visibles, plus ils correspondent à la superposition des uns et des autres.
Il est également intéressant de voir que les top évoluent et ne sont pas exactement les mêmes selon les années, certainement grâce à l'installation de nouveaux compteurs.
